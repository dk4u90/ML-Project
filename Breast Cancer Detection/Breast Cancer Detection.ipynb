```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_breast_cancer
cancer_dataset = load_breast_cancer()
type(cancer_dataset)
```

```python
print(cancer_dataset['filename'])
```

```python
cancer_df = pd.DataFrame(np.c_[cancer_dataset['data'],cancer_dataset['target']], columns=np.append(cancer_dataset['feature_names'],['target']))
```

```python
# create second DataFrame by droping target
cancer_df2 = cancer_df.drop(['target'],axis=1)
cancer_df2.shape
```

```python
# visualize correlation barplot
plt.figure(figsize=(16,5))
ax = sns.barplot(cancer_df2.corrwith(cancer_df.target).index,cancer_df2.corrwith(cancer_df.target))
ax.tick_params(labelrotation = 60)
```

```python
# input variable
x=cancer_df.drop(['target'], axis=1)
# output variable
y=cancer_df['target']
```

```python
# split dataset into train and test
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=5)
```

```python
# Feature scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train_sc = sc.fit_transform(x_train)
x_test_sc = sc.fit_transform(x_test)
```

```python
#Breast Cancer Detection Machine Learning Model Building
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
```

```python
#Support Vector Classifier
from sklearn.svm import SVC
svc_clasifier = SVC()
svc_clasifier.fit(x_train,y_train)
y_pred_scv = svc_clasifier.predict(x_test)
accuracy_score(y_test,y_pred_scv)
```

```python
svc_clasifier2 =SVC()
svc_clasifier2.fit(x_train_sc,y_train)
y_pred_scv_sc = svc_clasifier2.predict(x_test_sc)
accuracy_score(y_test,y_pred_scv_sc)
```

```python
# Logistic Regression
from sklearn.linear_model import LogisticRegression
lr_classifier = LogisticRegression()
lr_classifier.fit(x_train, y_train)
y_pred_lr = lr_classifier.predict(x_test)
accuracy_score(y_test, y_pred_lr)
```

```python
lr_calisifier2 = LogisticRegression()
lr_calisifier2.fit(x_train_sc,y_train)
y_pred_lr_sc = lr_calisifier2.predict(x_test_sc)
accuracy_score(y_test,y_pred_lr_sc)
```

```python
# K â€“ Nearest Neighbor Classifier
from sklearn.neighbors import KNeighborsClassifier
kc_classifier = KNeighborsClassifier()
kc_classifier.fit(x_train,y_train)
y_predict_kc = kc_classifier.predict(x_test)
accuracy_score(y_test,y_predict_kc)
```

```python
kc_classifier2 = KNeighborsClassifier()
kc_classifier2.fit(x_train_sc,y_train)
y_predict_kc_sc = kc_classifier2.predict(x_test_sc)
accuracy_score(y_test,y_predict_kc_sc)
```

```python
# Naive Bayes Classifier
from sklearn.naive_bayes import GaussianNB
nb_classifier = GaussianNB()
nb_classifier.fit(x_train,y_train)
y_pred_nb = nb_classifier.predict(x_test)
accuracy_score(y_test,y_pred_nb)
```

```python
nb_classifier2 = GaussianNB()
nb_classifier2.fit(x_train_sc,y_train)
y_pred_nb_sc = nb_classifier2.predict(x_test_sc)
accuracy_score(y_test,y_pred_nb_sc)
```

```python
#Confusion Matrix
cm = confusion_matrix(y_test,y_pred_nb)
sns.heatmap(cm,annot=True)
```

```python
#Classification Report of Model
print(classification_report(y_test, y_pred_nb))
```

```python
## Pickle
import pickle
 
# save model
pickle.dump(nb_classifier, open('breast_cancer_detector.pickle', 'wb'))
 
# load model
breast_cancer_detector_model = pickle.load(open('breast_cancer_detector.pickle', 'rb'))
 
# predict the output
y_pred = breast_cancer_detector_model.predict(x_test)
 
# confusion matrix
print('Confusion matrix of GaussianNB model: \n',confusion_matrix(y_test, y_pred),'\n')
 
# show the accuracy
print('Accuracy of GaussianNB model = ',accuracy_score(y_test, y_pred))
```
